```python
import pandas as pd
import numpy as np
```

```python
data_path = r"stock_price.pkl"
data = pd.read_pickle(data_path)
data[:5]
```

```python
daily_return = [row[3] for row in data]
daily_return[:5]
```

```python
flat_th = np.percentile(np.abs(daily_return),33)
print(flat_th)
```

```python
bd22_data_path = r"bd22_stock_price.pkl"
bd22_data = pd.read_pickle(bd22_data_path)
bd22_data[:5]
```

```python
close_price = pd.Series([row[2] for row in bd22_data])
bd22_daily_return = close_price/close_price.shift(1) -1
bd22_daily_return
```

```python
bd22_flat_th = np.percentile(np.abs(bd22_daily_return[1]),33)
print(bd22_flat_th)
```
